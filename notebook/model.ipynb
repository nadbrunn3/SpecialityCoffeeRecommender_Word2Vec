## standardizing columns

```python
def rename(df):
    """
    function renames columns of df
    input: dataframe
    output: a renamed dataframe    
    """
    
    col = df.columns.tolist()
    for idx, i in enumerate(col):
        if i in ['coffee_title', 'title']:
            col[idx] = 'coffee'
        elif i in ['coffee_country', 'coffee_origin', 'land']:
            col[idx] = 'origin'
        elif i in ['flavor', 'review_text', 'review', 'flavours']:
            col[idx] = 'flavour'
        elif i =='roast_level':
            col[idx] = 'roast'
        elif i in ['aufbereitung', 'production', 'Production']:
            col[idx] = 'process'      
        elif i == 'coffee_variety':
            col[idx] = 'variety'
        else: 
            col[idx] = i           
            
        df.columns = col
    return df
```

```python
df_list = [f1, f2, f3, f4, f5, f6, f7, f8, f9, f10]
key = ['19g', 'cf', 'cz', 'cr', 'cd', 'fel', ',gar', 'mis', 'red', 'ot']
df_renamed = []

# for the list of df apply the rename(x)-function and append all the renamed dfs into a new list:
for df in df_list:
    df = rename(df)
    df_renamed.append(df)

    

df_list_concat_coffee = []

# for all the renamed dfs only use the COLUMNS coffee, flavour
for df in df_renamed:
    df = pd.DataFrame(df)
    df1 = df[['coffee', 'flavour']]
    df_list_concat_coffee.append(df1)
```

## Normalizing and Vectorizing Coffee Flavour Profile

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

### Cleaning | Lemmatizer | Stopwords

```python

wordnet=WordNetLemmatizer()

normalized_corpus = []
normalized_sentences = []
for flav in df.flavour:
    sentences = nltk.sent_tokenize(flav)
    ml = []
    for i in range(len(sentences)):
        review = re.sub('[^a-zA-Z)]', ' ', sentences[i])
        review = re.sub('\W', ' ', sentences[i])
        review = review.lower()
        review = review.split()
        review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
        review = ' '.join(review)
        ml.append(review)
        normalized_sentences.append(review)
    normalized_corpus.append(' '.join(ml))
        
```

### Bi_Gram Tr_Gram

```python
from gensim.models.phrases import Phrases, Phraser
from gensim.models import Word2Vec
```

```python
sentence_stream = [doc.split() for doc in normalized_sentences]

bi_phrases = Phrases(sentence_stream, min_count = 1, threshold = 1 )
tri_phrases = Phrases(bi_phrases[sentence_stream])

# bi_grams = Phraser(bi_phrases)
ngrams = Phraser(tri_phrases)

phrased_sentences = []
for sent in sentence_stream:
    phrased_sentence = ngrams[sent]
    phrased_sentences.append(phrased_sentence)


words = [item for sublist in phrased_sentences for item in sublist]

words
```

### Embedding Coffee Flavour Profiles with Top 1000 

```python
def top_1000_only(word):
    if word in list(df_word_freq.index):
        return word
```

```python
# Bi Gram of each review in my database:

flavour_stream = [doc.split() for doc in normalized_corpus]

phrases = Phrases(flavour_stream)
phrases = Phrases(phrases[flavour_stream])

ngrams = Phraser(phrases)

phrased_flavours = []
for sent in flavour_stream:
    phrased_flavour = ngrams[sent]
    phrased_flavours.append(phrased_flavour)

```

```python
top_1000_descriptor_flavour_review = []  

for normalized_flavour in phrased_flavours:
    #for each review extract only the words that are present in the top 1000.
    descriptors = [top_1000_only(flav) for flav in normalized_flavour]
    no_nones = [str(d) for d in descriptors if d is not None]
    descriptorized_flavour_profil = ' '.join(no_nones)
    #save the reduced reviews in a list:
    top_1000_descriptor_flavour_review.append(descriptorized_flavour_profil)


    
```

### TF-IDF to account for more distinctive descriptors 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tidf = TfidfVectorizer(strip_accents = 'unicode', stop_words='english') 
tidf.fit(top_1000_descriptor_flavour_review)
```

```python
# # save model 

# import pickle

# pickle.dump(tidf, open('../models/tidf1000.pkl', 'wb'))
```

```python
feature_names = tidf.get_feature_names_out()
invers_doc_freq = tidf.idf_
```

```python
weights = dict(zip(feature_names, invers_doc_freq))
```

```python
# create a binary pickle file 
#f = open("../data_output/weights.pkl","wb")

# write the python object (dict) to pickle file
#pickle.dump(weights,f)

# close file
#f.close()
```

```python
vector_for_each_review = []
flav_count_for_each_review = []
list_of_flav_for_each_review = []

for descr in top_1000_descriptor_flavour_review:
    descr_count = 0
    weighted_vectors = []
    flav_list = descr.split()
    for flav in flav_list: 
        if flav in weights.keys():
            flav_idf = weights[flav]
            flav_w2v_vector = w2v_model.wv.get_vector(flav).reshape(1, 300)
            weighted_flav_vector = flav_idf * flav_w2v_vector
            weighted_vectors.append(weighted_flav_vector)
            descr_count += 1
        else: 
            continue
    
    try: 
        review_vector = sum(weighted_vectors)/len(weighted_vectors)
    except: 
        review_vector = []
        
    vector_for_each_review.append(review_vector)
    flav_count_for_each_review.append(descr_count)
    list_of_flav_for_each_review.append(flav_list)
        
    
        
            
```

### SpaCy - Find most similar word 

```python
def similar_word_in_list(input_word):
    import spacy
    nlp = spacy.load("en_core_web_lg")
    similarity = 0
    descriptor = ''
    input_word = nlp(input_word)
    for term in weights.keys():
        term = nlp(term)
        if input_word.similarity(term) > similarity:
            similarity = input_word.similarity(term)
            descriptor = str(term)
    return descriptor, similarity
        
most_similar_descriptor, sim = similar_word_in_list('sweet')
most_similar_descriptor, sim
```

### Match by Flavour

```python
def match_flavour_pick(input_flavours, number_of_suggestion = 10):
    
    weighted_input_vectors = []
        
    for flavour_term in input_flavours:
        flavour_term = wordnet.lemmatize(flavour_term).lower()     

        if flavour_term not in weights:
            most_similar_descriptor, similarity = similar_word_in_list(flavour_term)
            print('This coffee characteristic {} is not matched. Instead we will use the similar term {}'.format(flavour_term, most_similar_descriptor))
            print('- The similarity of the words is the highes we could find with {:.2f}'.format(similarity))
            print()            
            
            flav_idf = weights[most_similar_descriptor]
            flav_w2v_vector = w2v_model.wv.get_vector(most_similar_descriptor).reshape(1, 300)
            weighted_flav_vector = flav_idf * flav_w2v_vector
            weighted_input_vectors.append(weighted_flav_vector)
            continue

        else:
            try: 
                flav_idf = weights[flavour_term]
                flav_w2v_vector = w2v_model.wv.get_vector(flavour_term).reshape(1, 300)
                weighted_flav_vector = flav_idf * flav_w2v_vector
                weighted_input_vectors.append(weighted_flav_vector)
            except: 
                continue


    input_vector = sum(weighted_input_vectors)

    distance, indice = knn_model.kneighbors(input_vector, return_distance=True, n_neighbors=number_of_suggestion+1)
    distances = distance.tolist()[0][1:]
    indices = indice.tolist()[0][1:]
        
    recommended_coffee = []
    recommended_coffee_flavour_profil = []
    coffee_df_original = []
    coffee_df_original_index = []
    
    print('\n RECOMMENDED COFFEES: \n')

    n = 1
    for d, i in zip(distances, indices):
        coffee_df_original.append(coffee['level_0'][i])
        coffee_df_original_index.append(coffee['level_1'][i])
        suggested_coffee = coffee['coffee'][i]
        recommended_coffee.append(suggested_coffee)
        suggested_coffee_flavours = coffee['normalized_phrased_flavour'][i]
        recommended_coffee_flavour_profil.append(suggested_coffee_flavours)
        print('Suggestion', str(n), ': {} ---- with a cosine distance of {:.3f}'.format(suggested_coffee, d))
        print('This coffee has the following flavour profile: {}'.format(suggested_coffee_flavours))
        print('')
        n+=1
    
    return indices, recommended_coffee, recommended_coffee_flavour_profil, coffee_df_original, coffee_df_original_index

```

```python
my_flavours = ['rose', 'pomegranate']
indices, recommended_coffee, recommended_coffee_flavour_profil, coffee_df_original, coffee_df_original_index = match_flavour_pick(my_flavours)
```
