# Kaffeezentrale

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import random
import time
```

```python
# Get URL for all pages
pages = ['https://kaffeezentrale.at/de/kaffee/']
for i in range(2,7):
    pages.append('https://kaffeezentrale.at/de/kaffee/?page='+str(i))
pages
```

## GET COFFEE URLS

```python
coffee_url = []
for url in pages: 
    for i in range(60): 
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        coffee_url.append(soup.select('li.products__item > a')[i]['href'])
    
```

```python
coffee_url
```

```python
coffees = ['https://kaffeezentrale.at'+i for i in coffee_url]
```

```python
import numpy as np
```

## test

```python
coffee_title = []
review_text = []


for idx, url in enumerate(coffees):

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    try: 
        title = soup.select_one('section.details__description > h1').text
        coffee_title.append(title)
        text = soup.select_one('section.details__description > h5').text
        review_text.append(text)
    except:
        review_text.append(title)
        coffee_title.append(np.NaN)

    
    sleep = random.randint(1,2)    
    time.sleep(sleep)
    print('Coffee on idx {} parsed, sleeping for {} sec.'.format(idx, sleep) )
```

```python
my_dict = {'coffee_title': coffee_title,
           'review_text': review_text,
          'more_information': coffees}
```

```python
kaffeezentrale = pd.DataFrame.from_dict(my_dict)
```

```python
kaffeezentrale.to_csv('kaffeezentrale.csv')
```
