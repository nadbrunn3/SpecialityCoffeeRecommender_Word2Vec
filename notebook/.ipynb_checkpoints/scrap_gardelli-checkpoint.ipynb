# GARDELLI

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import random 
import time
```

```python
url = 'https://shop.gardellicoffee.com/14-coffees'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
link = []
for i in range(17):     
    link.append(soup.select('#js-product-list > div.products.row >article > div > a')[i]['href'])
    
link
```

```python
url = link[11]
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
my_dict = {'title': [],
           'rating': [],
           'herkunfstland_röster': [],
           'röster': [],
           'land': [],
           'region': [],
           'farm': [],
           'höhenlage': [],
           'variety': [],
           'aufbereitung': [],
           'review_text': []}
```

```python
#title 

soup.select_one('div.product-description > p').text.split('Notes')[1].replace('\xa0', '').split('Suggested')[0]
```

```python
soup.select('div.product-description> p > strong')
```

```python
for idx, url in enumerate(link):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')    
    
    # Title
    try: 
        t = soup.select_one('#main > div.row.itc_container > div.col-md-8 > h1').text
        my_dict['title'].append(t)
    except: 
        my_dict['title'].append(np.NaN)
        
    #rating
    try:        
        r = soup.select('dl.data-sheet > dd')[0].text
        my_dict['rating'].append(r)
    except: 
        my_dict['rating'].append(np.NaN)

    #röster
    my_dict['röster'].append('gardelli_speciality_coffees')

    #Kaffee Origin
    try: 
        h = soup.select('dl.data-sheet > dd')[3].text
        my_dict['land'].append(h)
    except: 
        my_dict['land'].append(np.NaN)

    # # Röstungsgrad --
    # try: 
    #     r = soup.select('div.spec-list > div > div')[9].text
    #     my_dict['röstung'].append(r)
    # except: 
    #     my_dict['röstung'].append(np.NaN)

    #Herkunfstland Röster --

    my_dict['herkunfstland_röster'].append('italy') 

    #Region
    try: 
        reg = soup.select('dl.data-sheet > dd')[4].text
        my_dict['region'].append(reg)
    except: 
        my_dict['region'].append(np.NaN)
        
    #Farm
    try: 
        f = soup.select('dl.data-sheet > dd')[2].text
        my_dict['farm'].append(f)
    except: 
        my_dict['farm'].append(np.NaN)

    #Höhenlage
    try: 
        lage = soup.select('dl.data-sheet > dd')[5].text
        my_dict['höhenlage'].append(lage)
    except: 
        my_dict['höhenlage'].append(np.NaN)

    #Varietät
    try: 
        v = soup.select('dl.data-sheet > dd')[7].text
        my_dict['variety'].append(v)
    except: 
        my_dict['variety'].append(np.NaN)

    #Aufbereitung
    try: 
        a = soup.select('dl.data-sheet > dd')[6].text
        my_dict['aufbereitung'].append(a)
    except: 
        my_dict['aufbereitung'].append(np.NaN)
        
    #Flavor
    try: 
        g = soup.select_one('div.product-description > p').text.split('Notes')[1].replace('\xa0', '').split('Suggested')[0]
        my_dict['review_text'].append(g)
    except:
        my_dict['review_text'].append(np.NaN)
        
    sleep = random.randint(1,2)
    time.sleep(sleep)
    
    print('finished with coffee {}, sleep for {} seconds'.format(idx, sleep))
```

```python
gardelli = pd.DataFrame.from_dict(my_dict)
```
