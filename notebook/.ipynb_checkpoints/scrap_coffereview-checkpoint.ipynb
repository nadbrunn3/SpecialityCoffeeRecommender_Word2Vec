```python
# Creating lists for all the information of interest: 

rating = []
roaster_name = []
coffee_title = []
roaster_location = []
coffee_origin = []
roast_level = []
agtron = []
price_estimate = []
review_date = []
aroma = []
acidity_structure = []
body = []
flavor = []
aftertaste = []
visit_roaster_link = []
blind_assessment_review_text = []

```

```python
for idx, link in enumerate(a_list_detailed_review):
    
    url = link
    response = requests.get(url, headers={'User-Agent': user_agent})
    
    if response.status_code != 200:
        print('Error while accessing webpage')
    else:
        soup = BeautifulSoup(response.content, "html.parser")
        print('fetching data for link idx {}'.format(idx))
    
    try:     
        # get rating
        r = soup.select_one('span.review-template-rating').text
        rating.append(r)
    except:
        rating.append('NaN')
        
    try:
        # get roaster name
        roaster = soup.select_one('p.review-roaster').text
        roaster_name.append(roaster)
    except:
        roaster.append('NaN')
        
    try:
        # get coffee brand title 
        brand = soup.select_one('h1.review-title').text
        coffee_title.append(brand)
    except:
        coffee_title.append('NaN')
        
    try:
        # get the roaster location
        location = soup.select('table.review-template-table > tr')[0].select('td')[1].text
        roaster_location.append(location)
    except:
        roaster_location.append('NaN')
        
    try:
        # get coffee origin
        origin = soup.select('table.review-template-table > tr')[1].select('td')[1].text
        coffee_origin.append(origin)
    except:
        coffee_origin.append('NaN')
        
    try:
        # get roast level
        rl = soup.select('table.review-template-table > tr')[2].select('td')[1].text
        roast_level.append(rl)
    except:
        roast_level.append('NaN')
        
    try:
        #get agtron
        agt = soup.select('table.review-template-table > tr')[3].select('td')[1].text
        agtron.append(agt)
    except:
        agtron.append('NaN')
    
    try:
        # get estimated price
        p = soup.select('table.review-template-table > tr')[4].select('td')[1].text
        price_estimate.append(p)
    except:
        price_estimate.append('NaN')
        
    try:
        # get review date
        date = soup.select('div.col-2 > table.review-template-table > tr')[0].select('td')[1].text
        review_date.append(date)
    except:
        review_date.append('NaN')

    try:
        # get aroma 
        a = soup.select('div.col-2 > table.review-template-table > tr')[1].select('td')[1].text
        aroma.append(a)
    except:
        aroma.append('NaN')
        
    try:
        # get acidity/structure
        acid = soup.select('div.col-2 > table.review-template-table > tr')[2].select('td')[1].text
        acidity_structure.append(acid)
    except:
        acidity_structure.append('NaN')
        
    try:
        # get body
        b = soup.select('div.col-2 > table.review-template-table > tr')[3].select('td')[1].text.split('\t')[0]
        body.append(b)
    except:
        body.append('NaN')

    try:
        # get flavor 
        fl = soup.select('div.col-2 > table.review-template-table > tr')[4].select('td')[1].text
        flavor.append(fl)
    except:
        flavor.append('NaN')
    
    try:
        # get aftertaste
        at = soup.select('div.col-2 > table.review-template-table > tr')[5].select('td')[1].text
        aftertaste.append(at)
    except:
        aftertaste.append('NaN')
     
    try:
        # get review_text
        text = soup.select_one('#genesis-content > article > div > div > p:nth-child(3)').get_text().split(':')[1].strip()
        blind_assessment_review_text.append(text)
    except:
        blind_assessment_review_text.append('NaN')    
    
    # respectful nap:
    wait_time = random.randint(1,2)
    time.sleep(wait_time)

```
