# Spezialitätenkaffee Coffeefriends

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import random 
import time
```

### Get Page Url

```python
pages = ['https://www.coffeefriend.at/c/spezialitatenkaffee/']

for i in range(2,7):
    pages.append('https://www.coffeefriend.at/c/spezialitatenkaffee/page/' + str(i))
```

```python
pages
```

### Get Coffee Url

```python
coffee_url = []
for url in pages:
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for i in range(12): 
        link = soup.select('body > div.wrap.container.woocommerce > div > main > ul > div > li > div > span')[i]['data-gtm4wp_product_url']
        coffee_url.append(link)
coffee_url
```

```python
len(coffee_url)
```

### Get Coffee Information 

```python
title= []
herkunfstland_röster= []
röster= []
land= []
region= []
farm= []
höhenlage= []
variety= []
aufbereitung= []
röstung= []
review_text= []
           
```

```python
for idx, url in enumerate(coffee_url):

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Title
    try: 
        t = soup.select_one('h1.single-product__title').get_text()
        my_dict['title'].append(t)
    except: 
        my_dict['title'].append(np.NaN)

    #röster
    try: 
        r = soup.select('div.spec-list > div > div')[5].text
        my_dict['röster'].append(r)
    except: 
        my_dict['röster'].append(np.NaN)

    #Kaffee Origin
    try: 
        h = soup.select('div.spec-list > div > div')[7].text
        my_dict['land'].append(h)
    except: 
        my_dict['land'].append(np.NaN)

    # Röstungsgrad
    try: 
        r = soup.select('div.spec-list > div > div')[9].text
        my_dict['röstung'].append(r)
    except: 
        my_dict['röstung'].append(np.NaN)

    #Herkunfstland Röster
    try: 
        hl = soup.select('div.spec-list > div > div')[11].text
        my_dict['herkunfstland_röster'].append(h)
    except: 
        my_dict['herkunfstland_röster'].append(np.NaN)

    #Region 
    try: 
        reg = soup.select_one('div.single-product__about__container__content > p:nth-child(14)').text
        my_dict['region'].append(reg)
    except: 
        my_dict['region'].append(np.NaN)
        
    #Farm
    try: 
        f = soup.select_one('div.single-product__about__container__content > p:nth-child(15)').text
        my_dict['farm'].append(f)
    except: 
        my_dict['farm'].append(np.NaN)

    #Höhenlage
    try: 
        lage = soup.select_one('div.single-product__about__container__content > p:nth-child(16)').text
        my_dict['höhenlage'].append(lage)
    except: 
        my_dict['höhenlage'].append(np.NaN)

    #Varietät
    try: 
        v = soup.select_one('div.single-product__about__container__content > p:nth-child(17)').text
        my_dict['variety'].append(v)
    except: 
        my_dict['variety'].append(np.NaN)

    #Aufbereitung
    try: 
        a = soup.select_one('div.single-product__about__container__content > p:nth-child(18)').text
        my_dict['aufbereitung'].append(a)
    except: 
        my_dict['aufbereitung'].append(np.NaN)
        
    #Flavor
    try: 
        g = soup.select_one('div.single-product__about__container__content > p:nth-child(20)').text
        my_dict['review_text'].append(g)
    except: 
        my_dict['review_text'].append(np.NaN)
        
    sleep = random.randint(1,2)
    time.sleep(sleep)
    
    print('finished with coffee {}, sleep for {} seconds'.format(idx, sleep))

coffee_friends = pd.DataFrame.from_dict(my_dict)
```

```python
def fill_list_with_nan(values):

    index = []
    text = []
    for i in values: 
        index.append(i[0])
        text.append(i[1])
        
    ml = []
    x = 0
    for idx, elem in enumerate(index):
        while x < 72:
            if elem == x: 
                ml.append(text[idx])
                x = x + 1
                break
            else: 
                ml.append(np.NaN)
                x = x + 1
    return ml  
    

        
```

```python
flavor = fill_list_with_nan(review_text)
roast = fill_list_with_nan(röstung)
production = fill_list_with_nan(aufbereitung)
variety = fill_list_with_nan(variety)
farm = fill_list_with_nan(farm)
region = fill_list_with_nan(region)

```
