```python
url = 'https://www.kaffeeothek.at/kaffeebohnen/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
kaffee_links = []
for i in range(73):
    kaffee_links.append(soup.select('div.product-tile-cont > div > a')[i]['href'])
```

```python
# Removing unwanted links: 

# del kaffee_links[3]
# del kaffee_links[6]

for idx, i in enumerate(kaffee_links): 
    if i in ['advent', 'probier']: 
        del kaffee_links[idx]
```

```python
my_dict = {'coffee': [],
          'region': [],
          'reg': [],
          'process': []}

for idx, link in enumerate(kaffee_links):
    url = link
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    try: 
        #Kaffee
        my_dict['coffee'].append(soup.select_one('div.et_pb_column.et_pb_column_1_3.et_pb_column_1.et_pb_css_mix_blend_mode_passthrough > div > h1').text)
    except:
        my_dict['coffee'].append('NaN')
        #process
    try: 
        my_dict['process'].append(soup.select_one('div.product-properties > div:nth-child(20)').text)
    except: 
        my_dict['process'].append('NaN')
            #region
    try: 
        my_dict['region'].append(soup.select_one('div.product-properties > div:nth-child(18)').text)
    except: 
        my_dict['region'].append('NaN')
        
    try: 
        my_dict['reg'].append(soup.select_one('div.product-properties > div:nth-child(16)').text)
    except: 
        my_dict['reg'].append('NaN')
```

```python
df = pd.DataFrame.from_dict(my_dict)
```

```python
import numpy as np
```

```python
df.loc[df.process == 'NaN', 'process'] = np.NaN
df.loc[df.region == 'NaN', 'region'] = np.NaN
df.loc[df.reg == 'NaN', 'reg'] = np.NaN
```

```python
df.to_csv('../data_raw/othek_process.csv', index=True)
```
