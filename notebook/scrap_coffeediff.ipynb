# CoffeeDiff Speciality Coffee Beans

```python
import requests
from bs4 import BeautifulSoup
import time
import random
import re
import pandas as pd
import numpy as np
```

```python
# Get URLs 

urls = ['https://coffeediff.co.uk/coffee-beans']

for i in range(2,35):
    urls.append('{}?page={}'.format(urls[0], i))

```

```python
urls
```

```python
response = requests.get(urls[0])
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
my_dict = {'coffee_title': [],
           'coffee_origin': [],
           'roaster_name': [],
           'variety': [],
           'production': [],
           'roast_level': [], 
           'flavour': [],         
          }
```

```python
for idx, url in enumerate(urls): 
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')   
    
    for i in range(100):

        # Title
        try: 
            title = soup.select('div.product-name')[i].select('div')[0].text
            my_dict['coffee_title'].append(title)
        except: 
            my_dict['coffee_title'].append(np.NaN)
            

        # Roaster
        try:
            r = soup.select('div.product-name')[i].select('div')[1].text
            my_dict['roaster_name'].append(r)
        except:
            my_dict['roaster_name'].append(np.NaN)

        # Origin
        try:
            origin = soup.select('div.product-detail.origin > ul > li')[i].text
            my_dict['coffee_origin'].append(origin)
        except: 
             my_dict['coffee_origin'].append(np.NaN)
           

        # Variety
        try: 
            x = soup.select('div.product-detail.variety')[i].text.replace('\n', '').split(':')[1].strip()
            my_dict['variety'].append(x)
        except:
            my_dict['variety'].append(np.NaN)

        # Process 
        try: 
            pr = soup.select('div.product-detail.process')[i].text.replace('\n', '').split(':')[1].strip()
            my_dict['production'].append(pr)
        except:
            my_dict['production'].append(np.NaN)

        # Roast Level 
        try:
            roast = soup.select('div.product-detail.roast')[i].text.replace('\n', '').split(':')[1].strip()
            my_dict['roast_level'].append(roast)
        except:
            my_dict['roast_level'].append(np.NaN)
            

        # flavour
        try:
            fl = soup.select('div.product-detail.flavour')[i].text.replace('\n', '').split(':')[1].strip()
            my_dict['flavour'].append(fl)
        except:
            my_dict['flavour'].append(np.NaN)
        
        
        ## Sleep 
        sleep = random.randint(1,2)
        time.sleep(sleep)
        print('page {}, coffee {} extracted'.format(idx, i))
```

```python
print(len(my_dict['coffee_title']))
print(len(my_dict['coffee_origin']))
print(len(my_dict['roaster_name']))
print(len(my_dict['variety']))
print(len(my_dict['production']))
print(len(my_dict['roast_level']))
print(len(my_dict['flavour']))
```

```python
coffeediff = pd.DataFrame.from_dict(my_dict)
```
